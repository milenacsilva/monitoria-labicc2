### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Lendo os arquivos

```python
df = pd.read_csv('dados/merge_full.csv')

by_size = [df.loc[df['size'] == 14580],
           df.loc[df['size'] == 29159],
           df.loc[df['size'] == 58318],
           df.loc[df['size'] == 116636]]

print(df)
```

### Definindo os pontos do plot

```python
x = list(df['size'].unique())

# Isso é analogo ao de baixo
# y = list()
# for x in by_size:
#     print(x)
#     y.append(np.mean(x['time']))

y = [np.mean(x['time']) for x in by_size]
e = [np.std(x['time']) for x in by_size]
```

### Dando nome para os eixos

```python
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')
```


### Plotando e estilizando

```python
plt.figure(figsize=[10, 10], facecolor='grey')
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')

plt.errorbar(x, y, yerr=e, fmt='o--', color='g', ecolor='r', capsize=3)
```

### Configurações da figura

```python
plt.figure(figsize=[10, 10], facecolor='grey')
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')

plt.ylim([0, 0.3])
plt.errorbar(x, y, yerr=e, fmt='o--', color='g', ecolor='r', capsize=3)
```

### Adicionando legendas

```python
plt.figure(figsize=[10, 10], facecolor='grey')
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')

plt.ylim([0, 0.3])

plt.errorbar(x, y, yerr=e, fmt='o--', color='g', ecolor='r', capsize=3, label='merge')
plt.legend()
```

### Salvando plots

```python
plt.figure(figsize=[10, 10], facecolor='grey')
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')

plt.ylim([0, 0.3])

plt.errorbar(x, y, yerr=e, fmt='o--', color='g', ecolor='r', capsize=3, label='merge')
plt.legend()
```
